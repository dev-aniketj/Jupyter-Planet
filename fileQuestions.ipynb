```python
# Q-1. WAP to draw a line from (2,7) to (4,10) using matplotlib?

# Answer-1.

import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([2, 4])
y_points = np.array([7, 10])
plt.plot(x_points, y_points)
plt.show()
```

```python
# Q-7. WAP to draw a multiple graphs using subplot?

# Answer-7.

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([2,4])
y1 = np.array([7,3])
plt.subplot(2,2,1)
plt.plot(x1, y1)

x2 = np.array([4,1])
y2 = np.array([10,12])
plt.subplot(2,2,2)
plt.plot(x2, y2)

x3 = np.array([5,3])
y3 = np.array([-2, 10])
plt.subplot(2,2,3)
plt.plot(x3, y3)

plt.show()
```

```python
# Q-8. WAP to draw a bar graph?

# Answer-8.

import matplotlib.pyplot as plt
import numpy as np

x = np.array([2,3,4,5,7])
y = np.array([1,7,5,2,3])

plt.bar(x,y, color='#ff0066')
plt.show()
```

```python
# Q-9. WAP to draw a pie chat graph?

# Answer-9.

import matplotlib.pyplot as plt
import numpy as np

x = np.array([3,1,2,5,2])

plt.pie(x, 
        labels=[ 'apple', 'orange', 'banana', 'watermelon', 'stawberry'], 
        colors=['#ff3333', 'orange', '#e6e600', '#47d147', '#ff4d94']
       )
plt.show()
```

```python
# Q-10. WAP to draw a scatter graph?

# Answer-10.

import matplotlib.pyplot as plt
import numpy as np

x = np.array([2,3,5,1,2,4])
y = np.array([10,6,-1,0,4,7])

plt.scatter(x,y)
plt.show()
```

```python
# Q-11. WAP to draw a area plot graph?

# Answer-11.

import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,3.5,4.2,5,6])
y = np.array([0.8,3,1.7,4,2,7,5])

plt.fill_between(x,y)
plt.show()
```

```python
# Q-12. WAP to draw a histogram graph?

# Answer-12.

import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,1,3,6])

plt.hist(y, bins=[1,2,3,4])
plt.show()
```

```python
# Q-13. WAP to draw multiple plots in a single graph? 

# Answer-13.

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([2,5])
x2 = np.array([4,8,9])
y1 = np.array([9,4.2])
y2 = np.array([10,9,7])

plt.plot(x1, y1, ls='--', color='blue')
plt.plot(x2, y2, color='red', marker='*')

plt.show()
```

```python
# Q-14. WAP to draw a multiple graphs using subplot with different styles?

# Answer-14.

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([2,4])
y1 = np.array([7,3])
plt.subplot(2,2,1)
plt.plot(x1, y1, color='green', ls='dashed')

x2 = np.array([4,1])
y2 = np.array([10,12])
plt.subplot(2,2,2)
plt.plot(x2, y2, color='red', marker='3')

x3 = np.array([5,3])
y3 = np.array([-2, 10])
plt.subplot(2,2,3)
plt.plot(x3, y3, color='orange', ls='dotted', marker='*')

plt.show()
```
