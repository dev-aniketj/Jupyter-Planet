```python
# Q-1. Draw a line in a diagram from (2,7) to (4,10) ?

import matplotlib.pyplot as plt
import numpy as np

# rect
x = np.array([3, 7, 7, 3, 3])
y = np.array([1, 1, 4, 4, 1])
plt.plot(x, y, color='blue')

# gate
x = np.array([5, 6, 6, 5, 5])
y = np.array([1, 1, 3, 3, 1])
plt.plot(x, y, color='blue')

# triangle
x = np.array([3, 7, 5, 3])
y = np.array([4, 4, 7, 4])
plt.plot(x, y, color='blue')

# window
x = np.array([3.5, 4.5, 4.5, 3.5, 3.5])
y = np.array([2, 2, 3, 3, 2])
plt.plot(x, y, color='red')

# window line 1 (horiz)
x = np.array([3.5, 4.5])
y = np.array([2.5, 2.5])
plt.plot(x, y, color='red', linestyle='dashdot')

# window line 2 (vert)
x = np.array([4, 4])
y = np.array([3, 2])
plt.plot(x, y, color='red', linestyle='dashdot')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

i = 1
j = 100

# initilize stage
x = np.array([i, j, j, i, i])
y = np.array([i, i, j, j, i])

while i <= j :    
    plt.plot(x, y)
    i = i + 2
    j = j - 2
    x = np.array([i, j, j, i, i])
    y = np.array([i, i, j, j, i])
    
plt.show()
```
